```python
#SVM_--SVM stands for Support Vector Machine, a supervised learning algorithm used for classification and regression tasks
#First step is importing the libraries

import pandas as pd
from sklearn.datasets import load_iris


iris=load_iris()
df=pd.DataFrame(data=iris,columns=iris.feature_names)

df.head(5)
```

```python
# iris

#Coverting the iris data to  dataframe

df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df.head()

#I will now add the target columns since we do not have it in the above dataframe

df["target"]=iris.target

df.head(20)
```

```python
#Mapping the target names so that where we have 0, we can have the target name
#I will do that by adding a column

df["Column_flower_name"]=df.target.apply (lambda x:iris.target_names[x])
df

```

```python
#wITH NOW THE ABOVE DATA WE CAN DO THE SVM_
#1- First I will check the length of the setosa data

len(df[df.target==0])

#2- secondly I will check the length of the versicolor data

len(df[df.target==1])

#3- Thirdly I will check the length of the virginica data
len(df[df.target==2])
```

```python
#Dividing this dataset into 3 datasets based on the 

df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
#I will now create a scatter plot on petals on df0
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.xlabel("petals length")
plt.ylabel("petals width")
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"])
plt.scatter(df1["petal length (cm)"],df0["petal width (cm)"], color="green", marker="+")
plt.scatter(df2["petal length (cm)"],df0["petal width (cm)"], color="black", marker="x")

```

```python
#Next step I will Train a Support Vector Machine to the classification-

from sklearn.model_selection import train_test_split
df.head()

x=df.drop(["target","Column_flower_name"],axis="columns")
y=df.target

print(x.head())
print(y.tail())
```

```python
#Divide data into tes and train set for x and y values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)#20% of the data will be used to test the model

len(x_train)
len(x_test)

len(y_train)
len(y_test)
```

```python
#Now I will create the SVModel

from sklearn.svm import SVC
model=SVC()
#Next step is training the model
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
